```python
fdf = pd.read_csv('spotify_data/data.csv')
fdf.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
## Change order of cols for my own sake and readability
new_order = ['name','artists','duration_ms','acousticness', 'speechiness','danceability','energy','tempo',
             'instrumentalness', 'liveness', 'loudness','mode', 'popularity','explicit','valence','key', 
             'release_date','year','id']
fdf = fdf.reindex(columns=new_order)
```

```python
fdf['artists'] = fdf['artists'].map(lambda x: x.strip('[\'').strip('\']'))
```

```python
from sklearn.neighbors import NearestNeighbors as nn
```

```python
nghbr = nn(n_neighbors=5)
nghbr.fit(traindf)
```

```python
def buildPlaylist(songIndex):
    distances, indicies = nghbr.radius_neighbors(X=[traindf.iloc[songIndex]],radius=5,return_distance=True,sort_results=True)

    rdf = pd.DataFrame(columns=fdf.columns)
    for i in indicies[0]:
        rdf = rdf.append(fdf.iloc[i],ignore_index=True)

    return rdf.head(10)
```

```python
#limit some more featues 
fdf.columns
tdfv2_cols = ['acousticness', 'speechiness',
       'danceability', 'energy', 'tempo', 'instrumentalness', 'liveness',
       'loudness', 'mode', 'popularity', 'explicit', 'valence', 'key','year']
tdfv2 = fdf[tdfv2_cols]

nghbrv2 = nn(n_neighbors=10)
nghbrv2.fit(tdfv2)
```

```python
def buildPlaylistV2(songIndex):
    distances, indicies = nghbrv2.kneighbors(X=[tdfv2.iloc[songIndex]],n_neighbors=10)

    rdf = pd.DataFrame(columns=fdf.columns)
    for i in indicies[0]:
        rdf = rdf.append(fdf.iloc[i],ignore_index=True)
    rdf['distance'] = pd.Series(distances[0], index=rdf.index)
    

    return rdf
```

```python
## make playlist based on inputed song, 
## AND allow user to select which audio features to include when searching
```

```python
def BuildCustomPlaylist(song,features=[]):
    
    custom_df = fdf[features]

    neighbor = nn(n_neighbors=10,algorithm='ball_tree',n_jobs=-1)
    neighbor.fit(custom_df)
    
    distances, indicies = neighbor.kneighbors(X=[custom_df.iloc[song]],n_neighbors=10)

    rdf = pd.DataFrame(columns=fdf.columns)
    for i in indicies[0]:
        rdf = rdf.append(fdf.iloc[i],ignore_index=True)
    rdf['distance'] = pd.Series(distances[0], index=rdf.index)
    
    return rdf
```
